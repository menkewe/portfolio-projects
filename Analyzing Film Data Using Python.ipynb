```python
# Import Libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # adjusts configuration of plots 

# Read in the Data

df = pd.read_csv(r'C:\Users\menke\OneDrive\Documentos\Portfolio Projects\project-4_python\Data\movies.csv')

# Looking at Data

df.head()
```

```python
# Check for Missing Data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Data Types for Columns

df.dtypes
```

```python
# Removing Rows With Null Cells

df = df.dropna()

# Update Column Data Type

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')

df.head()
```

```python
# Create Correct Year Column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df.head()
```

```python
# Show all rows
# pd.set_option('display.max_rows', None)

#Sort by gross high to low

df.sort_values(by=['gross'], inplace=False, ascending=False).head()
```

```python
# Show Distinct Values

df['company'].drop_duplicates().sort_values(ascending=False).head()
```

```python
# Remove Duplicates

# df.drop_duplicates()

df.head()
```

```python
# Scatter Plot - Budget vs. Gross
```

```python
plt.scatter(x=df['budget'], y=df['gross'])

plt.title('budget vs gross earnings')

plt.xlabel('budget for film')

plt.ylabel('gross earnings')

plt.show()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)

df.head()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Looking at correlation 

df.corr(method='pearson',numeric_only=True) #pearson, kendall, spearman
```

```python
# Show correlation matrix as heatmap

correlation_matrix = df.corr(method='pearson',numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matrix for numeric features')

plt.xlabel('movie features')

plt.ylabel('movie features')

plt.show()
```

```python
# looking at company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson',numeric_only=True)

sns.heatmap(correlation_matrix, annot=True)

plt.title('correlation matrix for numeric features')

plt.xlabel('movie features')

plt.ylabel('movie features')

plt.show()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
